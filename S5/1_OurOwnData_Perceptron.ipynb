```python
#Training----------
#1.Load the DATA(our own data)
x=[[1,3],[20,22],[1,2],[18,19],[5,6],[7,8]]
y=["boy","girl","girl","boy","girl","boy"]
#2.Load the MODEL(perceptron)
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.Fit DATA into MODEL
model.fit(x,y)
#Testing-----predict
op=model.predict([[20,21]])
print(op)
```
