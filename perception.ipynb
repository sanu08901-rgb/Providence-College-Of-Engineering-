```python
#Training----------
#1.Load the DATA(excel sheet)
import pandas as pd
file=pd.read_csv("/content/13 MLP winequality-red.csv")
x=file[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides"]]
y=file["quality"]
#Cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8)
#2.Load the MODEL(perceptron)
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.Fit DATA into MODEL
model.fit(xtrain,ytrain)
#validation
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy=",score*100)
#Testing-----predict
op2=model.predict([[1.5,2.5,3.5,4.5,5.1]])
print(op2)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm=confusion_matrix(ytest,op)
print(cm)
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
cmd.plot()
```
